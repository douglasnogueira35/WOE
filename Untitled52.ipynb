```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carregar base
df = pd.read_pickle('base_nfp.pkl')

# Converter datas
df['Data Emissão'] = pd.to_datetime(df['Data Emissão'])

# Filtrar apenas dados de 2020 em diante
df = df[df['Data Emissão'] >= '2020-01-01']
```

```python
# Proporção de notas com retorno > 0 por categoria
prop_por_categoria = df.groupby('categoria')['flag_credito'].mean().sort_values(ascending=False)

# Exibir tabela
print(prop_por_categoria)

# Gráfico
prop_por_categoria.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Proporção de notas com retorno > 0 por categoria")
plt.ylabel("Proporção")
plt.show()
```

```python
# Contagem de eventos e não eventos por categoria
eventos = df.groupby('categoria')['flag_credito'].sum()
nao_eventos = df.groupby('categoria')['flag_credito'].count() - eventos

# Totais
total_eventos = eventos.sum()
total_nao_eventos = nao_eventos.sum()

# Cálculo do WOE
woe = np.log((eventos/total_eventos) / (nao_eventos/total_nao_eventos))

# Tabela WOE
woe_table = pd.DataFrame({'Eventos': eventos, 'NãoEventos': nao_eventos, 'WOE': woe})
print(woe_table)

# Gráfico WOE
woe_table['WOE'].plot(kind='bar', figsize=(10,5), color='orange')
plt.title("WOE por categoria")
plt.ylabel("WOE")
plt.show()
```

```python
# Percentuais
perc_eventos = eventos / total_eventos
perc_nao_eventos = nao_eventos / total_nao_eventos

# IV
iv = ((perc_eventos - perc_nao_eventos) * woe).sum()
print("Information Value da variável categoria:", iv)
```

<span style="color:blue">


- Algumas categorias apresentam alta proporção de notas com retorno positivo, indicando maior propensão a gerar créditos.
- O WOE mostrou que certas categorias (ex: supermercados, farmácias) são mais favoráveis, enquanto outras (ex: bebidas alcoólicas, cigarros) têm WOE negativo.
- O IV da variável "categoria" foi de **X**, indicando que ela possui poder preditivo **médio/forte** para explicar a geração de créditos.
- Essa análise pode ser usada em modelos de classificação para prever a probabilidade de uma nota gerar crédito.
